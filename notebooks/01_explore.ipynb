```python
from pathlib import Path
import cv2
import pandas as pd
```

```python
raw_dir=Path("../data/raw")
real_videos = list((raw_dir / "original").glob("*.mp4"))
fake_videos=list((raw_dir / "Deepfakes").glob("*.mp4"))
```

```python
print(f"real count: {len(real_videos)}")
print(f"fake count: {len(fake_videos)}")
```

```python
print(type(real_videos[0]))
cap = cv2.VideoCapture(str(real_videos[0]))
```

```python
print("fps ", cap.get(cv2.CAP_PROP_FPS))
print("frame count ", cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
#cap.release()
```

```python
fourcc = int(cap.get(cv2.CAP_PROP_FOURCC))
codec = "".join([chr((fourcc >> 8*i) & 0xFF) for i in range(4)])
```
