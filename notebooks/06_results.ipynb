```python
import pandas as pd

clean_test_results_df = pd.read_csv("../results/clean_test_results.csv")
compressed_test_results_df = pd.read_csv("../results/compressed_test_results.csv")
validation_results_df = pd.read_csv("../results/validation_results.csv")

pd.set_option("display.max_colwidth",250)
```

```python
validation_results_df
```

```python
combined_test_df = clean_test_results_df[["model_name", "auc"]].merge(compressed_test_results_df[["model_name", "auc"]],
                                                                     on="model_name", suffixes=("_clean", "_compressed"))
```

```python
combined_test_df["gap"] = combined_test_df["auc_clean"] - combined_test_df["auc_compressed"]
```

```python
combined_test_df.sort_values("auc_clean", ascending=False)
```

```python

validation_results_df[["model_name" ,"auc", "real_f1", "fake_f1", "compression"]].sort_values("auc", ascending=False)
```

```python
combined_test_df = clean_test_results_df[["model_name", "auc","augment"]].merge(
            compressed_test_results_df[["model_name", "auc", "augment"]], on=["model_name", "augment"], suffixes=("_clean", "_compressed"))
```
